```python
import  random
board=list()
move=1

# Function to create an empty Tic Tac Toe board
def createboard():
    board = list()
    for i in range(0, 9):
        board.append(' ')
    return board

# Function to display the Tic Tac Toe board
def displayBoard(board):
    print(board[0]+' | '+board[1]+' | '+board[2])
    print('---------')
    print(board[3]+' | '+board[4]+' | '+board[5])
    print('---------')
    print(board[6]+' | '+board[7]+' | '+board[8])

# Function to handle player's move
def playermove(board,letter,playerno):  #'X' for player
    while True:
        try:
            position=int(input("Player "+str(playerno)+" input your next move ("+letter+") position (1-9) : "))
            # Check if the entered position is within the valid range and is empty
            if 1 <= position <= 9 and board[position-1]==' ':
                board[position-1]=letter
                break
            else:
                print("Position already occupied or invalid input! Try entering again :")
        except ValueError:
            print("Enter a valid move (a number between 1 and 9) : ")

# Function to check if a player has won
def iswon(board):
    # Check rows
    if (board[0] == board[1] == board[2] != ' ') or \
       (board[3] == board[4] == board[5] != ' ') or \
       (board[6] == board[7] == board[8] != ' '):
        return True
    # Check columns
    if (board[0] == board[3] == board[6] != ' ') or \
       (board[1] == board[4] == board[7] != ' ') or \
       (board[2] == board[5] == board[8] != ' '):
        return True
    # Check diagonals
    if (board[0] == board[4] == board[8] != ' ') or \
       (board[2] == board[4] == board[6] != ' '):
        return True

    return False

# Function to handle computer's move
def computermove(board,letter): #'O' for computer
    vacantplaces=list()
    move=0

    # Find all vacant places on the board
    for i in range(0,9):
        if board[i]==' ':
            vacantplaces.append(i)

    boardCopy = board[:]

    # Check if computer can win in the next move and take that move
    for i in vacantplaces:
        boardCopy[i]=letter
        if iswon(boardCopy):
            board[i]=letter
            print("Computer's move (O) at position (1-9) : "+str(i+1))
            move=1
            break
        else:
            boardCopy[i]=' '

    # Check if player can win in the next move and block them
    if move==0:
        for i in vacantplaces:
            boardCopy[i]='X'
            if iswon(boardCopy):
                board[i]=letter
                print("Computer's move (O) at position (1-9) : " + str(i+1))
                move=1
                break
            else:
                boardCopy[i]=' '

    # If no immediate win or block, prioritize center
    if move==0:
        if 4 in vacantplaces:
            board[4]=letter
            print("Computer's move (O) at position (1-9) : " + str(5))
            move=1

    # If center is not available, prioritize corners
    if move==0:
        corners = [0, 2, 6, 8]
        possible_moves = [i for i in corners if i in vacantplaces]
        if possible_moves:
            position = random.choice(possible_moves)
            board[position] = letter
            print("Computer's move (O) at position (1-9) : " + str(position + 1))
            move=1

    # If no other strategic moves, choose a random available spot
    if move==0:
        i=random.randrange(0,len(vacantplaces))
        board[vacantplaces[i]]=letter
        print("Computer's move (O) at position (1-9) : " + str(vacantplaces[i]+1))
        move=1

# Function to check if the board is full
def isfull(board):
    if board.count(' ')>=1:
        return False
    else:
        return True

# Main function to run the game
def main():
    print("Welcome to Tic Tac Toe !")
    print("'X' is your key and 'O' is computer's key in case you play against computer.\n")
    computer_score = 0
    player1_score = 0
    player2_score = 0
    player_score = 0
    match_tie=0
    ch='y'
    # Loop to play multiple games
    while ch.lower() == 'y':
        print("Enter 1 to play against player")
        print("Enter 2 to play against computer")
        # Get user's choice for game mode
        option = int(input("Enter your choice: "))
        print()
        # Game mode: Player vs Computer
        if option == 2:
            board=createboard()
            # Game loop for Player vs Computer
            while True:
                displayBoard(board)
                if not isfull(board):
                    if not iswon(board):
                        playermove(board,'X',1)
                        if not iswon(board):
                            if not isfull(board):
                                computermove(board,'O')
                            else:
                                match_tie+=1
                                print("Match Tie !")
                                break
                        else:
                            player_score=player_score+1
                            print("You Won... Good Job")
                            break
                    else:
                        computer_score = computer_score + 1
                        print("Computer won..... Nice Try!")
                        break
                else:
                    match_tie += 1
                    print("Match Tie !")
                    break
        # Game mode: Player vs Player
        elif option == 1:
            board=createboard()
            # Game loop for Player vs Player
            while True:
                displayBoard(board)
                if not isfull(board):
                    if not iswon(board):
                        playermove(board,'X',1)
                        if not iswon(board):
                            if not isfull(board):
                                displayBoard(board)
                                playermove(board,'O',2)
                            else:
                                match_tie += 1
                                print("Match Tie !")
                                break
                        else:
                            player1_score=player1_score+1
                            print("Player 1 Won... Good Job")
                            break
                    else:
                        player2_score=player2_score+1
                        print("Player 2 Won..... Good Job")
                        break
                else:
                    match_tie += 1
                    print("Match Tie !")
                    break
        else:
            print("Invalid input: ")
        # Ask if the user wants to play again
        ch=input("Want to play again (y/any key) : ")
    # Display final scores
    print()
    print("Final Scores are : ")
    print("===================")
    print()
    print(" : Against computer : ")
    print("----------------------")
    print("Computer Score = "+str(computer_score)+"         Your Score = "+str(player_score))
    print(" : Player Against Player : ")
    print("---------------------------")
    print("Player 1 Score = " + str(player1_score)+"        Player 2 Score = " + str(player2_score))
    print("--------------------")
    print("Match Tied = "+str(match_tie))

# Start the game
main()
```
